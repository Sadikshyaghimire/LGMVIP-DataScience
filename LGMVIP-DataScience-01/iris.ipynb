```python

name = ["sepal width in cm" , "petal length in cm" , "petal width in cm" , "class"]
   
```

```python
df = pd.read_csv ("iris\iris.data", header=None, names=name)
df
```

# Exploratory Data Analysis


```python

skew_value = df['sepal width in cm'].skew()
print(f"Skewness of 'sepal width in cm': {skew_value}")
if skew_value > 0:
    print("The data is right-skewed (positively skewed).")
elif skew_value < 0:
    print("The data is left-skewed (negatively skewed).")
else:
    print("The data is approximately symmetric.")
```

```python
import seaborn as sns
sns.pairplot(df, hue='class')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='class', y='sepal width in cm')
plt.title('Box Plot of Sepal Width by Class')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
for species in df['class'].unique():
    sns.histplot(df[df['class'] == species]['sepal width in cm'], label=species, kde=True)
plt.title('Histogram of Sepal Width by Class')
plt.xlabel('Sepal Width in cm')
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have loaded your dataset into a DataFrame named 'iris_data'
# Select the numerical columns for correlation analysis
numerical_columns = ['sepal width in cm', 'petal length in cm', 'petal width in cm']

# Calculate the correlation matrix
correlation_matrix = df[numerical_columns].corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have loaded your dataset into a DataFrame named 'iris_data'
# Replace 'class' with the actual column name containing the class labels
class_column = 'class'

# Count the number of occurrences of each class
class_counts = df[class_column].value_counts()

# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=class_counts.index, y=class_counts.values, palette="viridis")
plt.xlabel('Iris Species')
plt.ylabel('Count')
plt.title('Class Distribution (Bar Plot)')
plt.xticks(rotation=45)
plt.show()

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', colors=sns.color_palette("Set3"))
plt.title('Iris Flower Class Distribution (Pie Chart)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have loaded your dataset into a DataFrame named 'iris_data'
# Replace 'class' with the actual column name containing the class labels
class_column = 'class'

# Replace these with the actual names of the numerical features
numerical_features = ['sepal width in cm', 'petal length in cm', 'petal width in cm']

# Create subplots for class distribution and feature distributions
fig, axes = plt.subplots(nrows=2, ncols=len(numerical_features), figsize=(16, 8))

# Class Distribution (Bar Plot)
class_counts = df[class_column].value_counts()
sns.barplot(x=class_counts.index, y=class_counts.values, ax=axes[0, 0], palette="viridis")
axes[0, 0].set_xlabel('Iris Species')
axes[0, 0].set_ylabel('Count')
axes[0, 0].set_title('Class Distribution (Bar Plot)')

# Create histograms and density plots for each numerical feature within each class
for i, feature in enumerate(numerical_features):
    for j, class_label in enumerate(df[class_column].unique()):
        subset = df[df[class_column] == class_label]
        
        # Create histograms
        sns.histplot(subset[feature], bins=20, kde=True, ax=axes[1, i], label=class_label)
        
        # Set labels and titles for feature distributions
        axes[1, i].set_xlabel(feature)
        axes[1, i].set_ylabel('Frequency')
        axes[1, i].set_title(f'Distribution of {feature}')
        
        # Add legend to the last subplot
        if i == len(numerical_features) - 1:
            axes[1, i].legend(title=class_column)

# Adjust layout and display
plt.tight_layout()
plt.show()
```
